```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return abs(x)
```

```python
def f2(x):
    return x * np.sin(5*x)
```

```python
def f3(x):
    ret = []
    for i in x:
        if i > 0:
            ret.append(i + 0.2 * np.sin(5*i))
        else:
            ret.append(0.2 * np.sin(5*i))
    return np.array(ret)
```

```python
def f4(x1, x2):
    y1 = np.abs(x1 + x2)
    y2 = np.abs(x1 - x2)
    return [[y1], [y2]]
```

```python
def sample(f):
    st = -np.sqrt(3)
    ed = np.sqrt(3)
    x = np.random.uniform(st, ed, size=300)
    y = f(x)
    plt.scatter(x, y, s=1)
    plt.show()
```

```python
def sample2(f):
    st = -np.sqrt(3)
    ed = np.sqrt(3)
    x1 = np.random.uniform(st, ed, size=300)
    x2 = np.random.uniform(st, ed, size=300)
    y = f(x1, x2)
    return y
    #plt.scatter(x, y, s=1)
    #plt.show()
```

```python
sample(f1)
```

```python
sample(f2)
```

```python
sample(f3)
```

```python
sample2(f4)
```
